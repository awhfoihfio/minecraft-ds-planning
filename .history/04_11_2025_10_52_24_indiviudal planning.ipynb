Data Description
This project uses data collected from a research Minecraft server operated by UBC Computer Science. The data consist of two tables: players.csv and sessions.csv. The players.csv file contains one row per unique player, while the sessions.csv file records individual gameplay sessions for those players.
The players.csv dataset includes 7 variables and 196 players.
The key variables are:
experience(): Self-reported Minecraft skill level (Beginner/Amateur/Regular/Veteran/Pro)
subscribe: Whether the player subscribed to the research newsletter (TRUE/FALSE)
hashedEmail: Anonymous player identifier
played_hour s:Total hours played on the server
name: Player name 
gender: Player gender
Age:Player reported age


The sessions.csv dataset contains 5 variables and 1535 rows. Each row represents one play session. It includes hashed player ID, session start and end times, and timestamp versions of those times.
